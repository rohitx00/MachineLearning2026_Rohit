## Numpy indexing and slicing of matrix

```python
arr = np.arange(1,21).reshape(4,5)
arr
```

```python
arr[3,4] # accessing element at 4th row and 5th column
```

```python
arr[0:2, 1:3] # slicing 2D array
```

## Array operation

```python
import numpy as np
arr1 = np.arange(1, 11)
arr2 = np.arange(11, 21)
```

```python
sum = arr1 + arr2
print(sum)
```

```python
minus = arr2 - arr1
print(minus)
```

```python
multiple = arr1 * arr2
print(multiple)
```

```python
div = arr2 / arr1
print(div)
```

```python
floor_div = arr2 // arr1
print(floor_div)
```

```python
power = arr2 ** arr1
print(power)
```

## Broadcasting

```python
l = [1, 2, 3, 4, 5]
arr = np.array(l)
```

```python
# adding 10 to each element of the list using loop
for num in l:
    num = num + 10
    print(num)
```

```python
arr + 10 # adding 10 to each element of the array using broadcasting
```

## Advance array manipulation

### Stacking array

```python
np.hstack((mat1,mat2)) # order of the row should be same
```

```python
np.vstack((mat1,mat2)) # order of the column should be same
```
