## np array vs python list
- datatypes - same data types only vs mixed allowed
- performance - much faster vs slower
- memory efficiency - high vs low
- vector operations - Easy + fast vs manual loops
 

```python
import numpy as np 
```

## Basic Array creation

```python
arr = np.array([1, 2, 3])
print(arr)
```

```python
d2_arr = np.array([[1, 2, 3], [4, 5, 6]])
print(d2_arr)
```

## Generating Array

```python
arr = np.arange(0, 10, 2)
print(arr)
```

```python
arr = np.zeros((2, 3))
print(arr)
```

```python
arr = np.ones((3, 4))
print(arr)
```

```python
arr = np.linspace(1,5,20)
print(arr)
```

## Random number Generation

```python
arr = np.random.rand(5) # output will be between 0 and 1
print(arr)
```

```python
arr = np.random.randn(5) # output will be between -3 and 3
print(arr)
```

```python
arr = np.random.randint(0, 10, (3, 3))
print(arr)
```

## Reshaping and resizing


```python
arr = np.arange(1,31)
arr
```

## Numpy indexing and slicing of vectors

```python
arr = np.arange(1,31)
arr
```

```python
arr[9] # accessing 10th element using indexing
```

```python
arr[1:6:1] #[start:stop:step] slicing
```

## Numpy indexing and slicing of matrix

```python
arr = np.arange(1,21).reshape(4,5)
arr
```

```python
arr[3,4] # accessing element at 4th row and 5th column
```

```python
arr[0:2, 1:3] # slicing 2D array
```

```python
slice = arr[1:4, 3:5] # slicing 2D array, this will create a view of the original array, not a copy
print(slice)
```

## Boolean indexing

```python
bool_index = arr % 2 != 0 # boolean indexing to get odd numbers
print(bool_index)
```

```python
arr = arr[bool_index] # using boolean indexing to get odd numbers from the array
print(arr)
```
